```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Generate a synthetic time series with some trend and noise
np.random.seed(42)
time = pd.date_range(start='2020-01-01', periods=200, freq='D')
trend = np.linspace(10, 50, 200)
for i in range(0,len(trend),10):
    trend[i]=50
noise = np.random.normal(loc=0, scale=5, size=200)
data = trend + noise

# Create a DataFrame for the time series
ts = pd.Series(data, index=time)

# Plot ACF and PACF
plt.figure(figsize=(12, 6))

# ACF plot
plt.subplot(121)
plot_acf(ts, lags=30, ax=plt.gca())
plt.title('Autocorrelation Function (ACF)')

# PACF plot
plt.subplot(122)
plot_pacf(ts, lags=30, ax=plt.gca())
plt.title('Partial Autocorrelation Function (PACF)')

plt.tight_layout()
plt.show()
```
